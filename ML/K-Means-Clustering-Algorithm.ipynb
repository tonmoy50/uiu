```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import datasets
import random
```

```python
iris = datasets.load_iris()
X = iris.data[:, :2]
length = X.shape[0]
#print(length)
y = []
for i in range(length):
    y.append([0, 0])
#print(y.shape[0])

k=3
CentreIndex = list(random.sample(range(0, length), k))
print(CentreIndex)

Centres = []
for i in range(k):
    Centres.append(list(X[CentreIndex[i]]))
    
print(Centres)
```

```python
for i in range(5) :
    print(X[i,:])
    
```

```python
counter = 0
while counter<100 :
    distances = np.zeros((length, 3))
    lengths = []
    for i in range(k):
        lengths.append(0)

    for i in range(length):
        for j in range(k):
            distances[i][j] = ((Centres[j][0] - X[i][0])**2 + (Centres[j][1] - X[i][1])**2)**(0.5)
        
    print('Distances:')
    for i in range(5):
        print(distances[i])
        
    for i in range(length):
        minimum = distances[i][0]
        minIndex = 0
        for j in range(k):
            if minimum>distances[i][j]:
                minimum = distances[i][j]
                minIndex = j
        
        #print('Centres:')
        #rint(Centres[minIndex])
        y[i] = list(Centres[minIndex])
        lengths[minIndex] += 1
    
    print('y:')
    for i in range(5):
        print(y[i])
    
    lossDistance = [0, 0]
    for i in range(length):
        lossDistance[0] += y[i][0] - X[i][0]
        if lossDistance[0]<0 :
            lossDistance[0] *= -1
        
        lossDistance[1] += y[i][1] - X[i][1]
        if lossDistance[1]<0 :
            lossDistance[1] *= -1
            
    print('Loss Distance: ')
    print(lossDistance[0] + lossDistance[1])
    
    avg = np.zeros((k, 2))

    for i in range(length):
        index = 0
        for j in range(k):
            if y[i] == Centres[j]:
                index = j
                #print('Index: ')
                #print(index)
                #print('Value: ')
                #print(y[i])
                #print('Centre: ')
                #print(Centres[j])
    
        avg[index][0] += X[i][0]
        avg[index][1] += X[i][1]

    flag = 0
    for i in range(k):
        avg[i][0] /= lengths[i]
        avg[i][0] = round(avg[i][0], 2)
        avg[i][1] /= lengths[i]
        avg[i][1] = round(avg[i][1], 2)
        if avg[i][0] != Centres[i][0] or avg[i][1] != Centres[i][1] :
            flag = 1
            
   
    #print('Average:')
    #print(avg)
    print(flag)

    if flag==1:
        for i in range(k):
            Centres[i][0]=round(avg[i][0], 2)
            Centres[i][1]=round(avg[i][1], 2)
    else:
        print('Centres:')
        print(Centres)
        break
    
    print('Centres:')
    print(Centres)
    
    counter += 1
```
